### **Mutli-Linear Regression**

```python
df.head()
```

#### **Encoding of data**
- binary 
- on hot encoder

```python
df_copy = df # making a copy of the original data
df_copy =  pd.get_dummies(df_copy, columns=['furnishingstatus'], drop_first=True)
```

```python
bin_cat_columns =[
    'mainroad',	'guestroom','basement',	'hotwaterheating',	'airconditioning',	'prefarea'
] 

for col in bin_cat_columns:
    df_copy[col] = df_copy[col].map({'yes':1, 'no':0})
```

```python
df_copy.head(1)
```

```python
X = df_copy.drop('price', axis=1)
y = df_copy['price'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model = linear_model.LinearRegression()
model
```

```python
model.fit(pd.DataFrame(X_train),pd.DataFrame(y_train))
```

```python
y_pred = model.predict(X_test)
```

```python
# mean_squared_error, r2_score
MSE = mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)
r_squared = r2_score(y_test, y_pred)

print(f"MSE: {MSE}") # 0 - infinity => low is good, my model is off by this value
print(f"RMSE: {RMSE}") # 0 - infinity => low is good, my model is off by this value
print(f"R^2: {r_squared}") # -ve infinity towards 1 =>  higher is better
```

##### **Summary:**
- our model explinas 65% of the data well, hwile its predictions are off by 1209843.801867475
